```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier


# Använder data från textfilen
data = np.loadtxt("datapoints.txt", delimiter=",", skiprows=1)

# Splittar data till features och label
X_train = data[:, :2]  
y_train = data[:, 2]   


# Klassifierar skillnaden mellan pichu och pikachu
def classify_test_point(test_point, k=1):

    if any(value < 0 for value in test_point):
        raise ValueError("test_point input contains a negative number")
  

    knn = KNeighborsClassifier(n_neighbors=k)
    
 
    knn.fit(X_train, y_train)
    
    
    prediction = knn.predict([test_point])
 

    if prediction == 0:
        print(f"Test point {test_point} is classified as Pichu.")
    else:
        print(f"Test point {test_point} is classified as Pikachu.")




# Skriver in en test punkt för att få reda på om den punkten är en pikachu eller pichu
test_point = [20.5, 34]  
classify_test_point(test_point)



# definierar plot funktionerna
def plot_test_point(test_point):
    plt.scatter(test_point[0], test_point[1], color="blue", label="Test Point", marker="x", s=100)
    plot_data(X_train, y_train)



def plot_data(X, y):
    plt.figure(figsize=(8, 6))
    for label in np.unique(y):
        plt.scatter(X[y == label][:, 0], X[y == label][:, 1], label=f"Class {int(label)}")
    plt.title("Pichu (blue) vs Pikachu (orange) Classification")
    plt.xlabel("Width (cm)")
    plt.ylabel("Height (cm)")
    plt.legend()
    plt.show()

# gör så att test punkterna kommer upp i en graf plot
plot_test_point(test_point)
```
